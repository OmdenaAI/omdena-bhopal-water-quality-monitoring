# Bar Graph 

```python
df['ndwi_perc'] = -df['ndwi'] * 100

fig = go.Figure(go.Bar(
    x=df['date'],
    y=df['ndwi_perc'],
    marker_color=df['ndwi_perc'],
    text=df['ndwi_perc'].round(2),
    textposition='outside'
))

fig.update_layout(
    title='NDWI Values',
    xaxis_title='Date',
    yaxis_title='NDWI (%)',
    template='plotly_dark'
)

fig.show()
```

```python
df['mndwi_perc'] = -df['mndwi'] * 100

fig = go.Figure(go.Bar(
    x=df['date'],
    y=df['mndwi_perc'],
    marker_color=df['mndwi_perc'],
    text=df['mndwi_perc'].round(2),
    textposition='outside'
))

fig.update_layout(
    title='MNDWI Values',
    xaxis_title='Date',
    .
    yaxis_title='MNDWI(%)',
    template='plotly_dark'
)

fig.show()
```

```python
df['ndci_perc'] = -df['ndci'] * 100

fig = go.Figure(go.Bar(
    x=df['date'],
    y=df['ndci_perc'],
    marker_color=df['ndci_perc'],
    text=df['ndci_perc'].round(2),
    textposition='outside'
))

fig.update_layout(
    title='NDCI Values',
    xaxis_title='Date',
    yaxis_title='NDCI(%)',
    template='plotly_dark'
)

fig.show()
```

This code generates a bar graph of the modified normalized difference water index (MNDWI) values over time. The MNDWI is another commonly used index for water detection in remote sensing, which is calculated using green and shortwave infrared (SWIR) bands of satellite imagery. The MNDWI values range from -1 to 1, where higher values indicate more water content.
Similar to the NDWI graph, the MNDWI graph shows that the water content in the area fluctuates over time. The MNDWI values range from -0.5 to 0.8, with higher values indicating a higher water content. The highest MNDWI values appear to be around June 2020 and February 2021, while the lowest values appear to be around September 2020.

The MNDWI graph provides additional information about the water content in the area, which can be useful for various applications such as monitoring changes in water levels and assessing the impact of drought or floods.

Overall, the ndci values seem to be somewhat lower than the ndwi and mndwi values, indicating that there may be more vegetation in the area than water. However, there are still periods where the ndci values drop significantly, suggesting that water is still an important factor in the region.

**How did converting to percentage help in analysis?**

Converting the NDCI, MNDWI, and NDWI values to percentages helped in creating the graph by scaling the values to a common range (-100 to 0). This allowed for easier comparison and interpretation of the results across the different dates.

For example, without converting the values to percentages, the NDCI values might have ranged from -0.3 to 0.4, while the MNDWI values ranged from -0.8 to 0.2. By converting both of these sets of values to percentages, they are now on the same scale (-100% to 0%), making it easier to compare and interpret the results.

Additionally, by multiplying the values by -100, the resulting percentages are now in a more intuitive range where a higher value indicates a more positive outcome (i.e. higher water content for NDWI, lower impervious surfaces for MNDWI, and healthier vegetation for NDCI).
